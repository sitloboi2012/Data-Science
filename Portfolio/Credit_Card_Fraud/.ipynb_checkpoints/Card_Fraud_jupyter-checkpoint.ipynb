```python
#Read data
data = pd.read_csv("creditcard.csv",sep=",")

#Get basic info
#Head data
data.head()
```

```python
#Count number of Class
#We have 2 class in this case - 1 is FRAUD and 2 is NORMAL
#Create count + table to show out (use histogram + sns.countplot)

fraud_total = 0
normal_total = 0
for i in data["Class"]:
    if i == 1:
        fraud_total += 1
    elif i == 0:
        normal_total += 1
    else:
        pass

ax = sns.countplot(x="Class",data=data)
print("Sum of Fraud Value is:",fraud_total)
print("Sum of Normal Value is:",normal_total)
```

```python
#Create table for Fraud and Normal - use that to creat graph and comparison
#Firstly look at the correlation if the relationship is strong or not 
corr = data.corr()
ax = sns.heatmap(corr, xticklabels=corr.columns,yticklabels=corr.columns)
```

```python
target = data["Class"]
ax = data.hist(bins=100)
for axis in ax.flatten():
    axis.set_xticklabels([])
    axis.set_yticklabels([])

plt.show()
```
