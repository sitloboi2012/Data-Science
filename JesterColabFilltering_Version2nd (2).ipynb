```
mean_rating = pd.DataFrame(train_data.groupby("joke_id")["rating"].mean())
```

```
mean_rating
```

```
mean_rating["Count"] = train_data.groupby("joke_id")["rating"].count()
mean_rating.sort_values(by="Count")
```

```
joke_user = train_data.pivot_table(index="user_id", columns="joke_id", values="rating")
```

```
joke_user.head()
```

```
joke_id = int(input("Put in a joke_id: "))
corr = joke_user.corrwith(joke_user[30])
recommend = pd.DataFrame(corr, columns=["Correlation"])
recommend.dropna(inplace=True)
recommend = recommend.join(mean_rating["Count"])
recommend = recommend[recommend["Count"] > 1000].sort_values("Correlation", ascending=False).reset_index()
list_top = recommend["joke_id"].head().tolist()
list_corr = recommend["Correlation"].head().tolist()

print("Because you like joke_ id {}".format(joke_id))
print()
print("How about try this these joke id: {}".format(list_top))
```
